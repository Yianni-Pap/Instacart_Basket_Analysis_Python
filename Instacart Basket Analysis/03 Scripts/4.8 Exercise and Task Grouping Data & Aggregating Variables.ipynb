## Aggregating Data with transform()

```python
df_orders_products_merged['max_order'] = df_orders_products_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_orders_products_merged.head(15)
```

## Deriving Columns with loc()


```python
# Creating a flag
df_orders_products_merged.loc[df_orders_products_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df_orders_products_merged.loc[(df_orders_products_merged['max_order'] <= 40) & (df_orders_products_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df_orders_products_merged.loc[df_orders_products_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
df_orders_products_merged['loyalty_flag'].value_counts(dropna=False)
```

```python
# QN 5 -- Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer)
df_orders_products_merged.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max', 'sum']})
```

The prices of products purchased by loyal customers differ from those purchased by regular or new customers. The loyal customers spend the least in terms of average price per item and they contribute to the total revenue leass than the regular customers. 
The new customers are buying on average the most expensive items.

```python
# QN 6 -- First we need to create a new column with the average price
df_orders_products_merged['avg_price']=df_orders_products_merged.groupby(['user_id'])['prices'].transform(np.max)
```

```python
df_orders_products_merged['avg_price']
```

```python
df_orders_products_merged.head(5)
```

```python
# Now we need to create the spending_flag
df_orders_products_merged.loc[df_orders_products_merged['avg_price'] < 10, 'spending_flag']='Low spender'
df_orders_products_merged.loc[df_orders_products_merged['avg_price'] >= 10, 'spending_flag']='High spender'
```

```python
# Here we count the values for the new flag
df_orders_products_merged['spending_flag'].value_counts(dropna=False)
```

```python
# Here we check the results
df_orders_products_merged.head(15)
```

```python
# QN 7 -- We need to use the same process as above
df_orders_products_merged['median_order']=df_orders_products_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.max)
```

```python
df_orders_products_merged['median_order']
```

```python
df_orders_products_merged.loc[df_orders_products_merged['median_order'] > 20, 'order_frequency_flag']='Non-frequent customer'
df_orders_products_merged.loc[(df_orders_products_merged['median_order'] > 10)&(df_orders_products_merged['median_order'] <= 20), 'order_frequency_flag']='Regular customer'
df_orders_products_merged.loc[df_orders_products_merged['median_order'] <= 10, 'order_frequency_flag']='Frequent customer'
```
